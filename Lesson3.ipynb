```python
t = np.linspace(0, 2*np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
rho = np.sqrt(x**2 + y**2)
phi = np.arctan2(y,x)*180/np.pi
plt.polar(phi, rho)
```

### 4. Задание 
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x)-1)/x+1)
plt.plot(x, x**2 -1)
plt.ylim(-1,7)
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def eq(p):
    x,y = p
    return(y - x**2 +1, np.exp(x) + x - x*y -1)
x1, y1 = fsolve(eq,(-2,1))
print(x1,y1)

x2,y2 = fsolve(eq,(2,2))
print(x2,y2)
```

Решите систему уравнений и неравенств:
y = x2 – 1

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, np.exp(x) + x*(2 - x**2) -1)
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def eq(x):
     return(np.exp(x) + x*(2 - x**2) -1)
x1 =  fsolve(eq,-2)
print(x1)
x2 =  fsolve(eq,2)
print(x2)
x0 =  fsolve(eq, .1)
print(x0)
```
